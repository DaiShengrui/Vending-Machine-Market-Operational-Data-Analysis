# 数据读取与处理

```python
import pandas as pd
data = pd.read_csv('/home/mw/input/data4910/vending_machine_sales.csv')
df = data.rename(columns={'Location':'位置','Machine':'用户友好的机器名称','Product':'从机器售出的产品','Category':'类型','Transaction':'每次交易的唯一标识符','TransDate':'交易的日期和时间',
                          'Type':'交易类型(现金/信用)','RCoil':'用于售出产品的线圈号#','RPrice':'价格','RQty':'已售数量','MCoil':'用于售出产品的映射线圈号','MPrice':'映射价格',
                          'MQty':'已售映射数量','LineTotal':'每次交易的总销售','TransTotal':'交易的总计','Prcd Date':'处理交易的日期'})
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna()
```

```python
df = df.drop(['Status','Device ID','每次交易的唯一标识符','用于售出产品的线圈号#','用于售出产品的映射线圈号'],axis=1)
```

```python
#改为日期类型
df['交易的日期和时间'] = pd.to_datetime(df['交易的日期和时间'], format='%m/%d/%Y')
df['处理交易的日期'] = pd.to_datetime(df['处理交易的日期'], format='%m/%d/%Y')
df.head()
```
